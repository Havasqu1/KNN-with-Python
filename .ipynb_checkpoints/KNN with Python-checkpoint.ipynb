# K-Nearest Nieghbors with Python

```python
import pandas as pd # data processing
from termcolor import colored as cl # elegant printing of text
import seaborn as sb # visualizations
import matplotlib.pyplot as plt # editing visualizations
from matplotlib import style # setting styles for plots
from sklearn.preprocessing import StandardScaler # normalizing data
from sklearn.neighbors import KNeighborsClassifier # KNN algorithm
from sklearn.metrics import accuracy_score # algorithm accuracy
from sklearn.model_selection import train_test_split # splitting the data

style.use('seaborn-whitegrid')
plt.rcParams['figure.figsize'] = (20, 10)
```

```python
# Importing Data

df = sb.load_dataset('iris')
print(cl(df, 'red', attrs=['bold'])) #exploring colored package
```

```python
# Data Description

print(cl(df.describe(), attrs = ['bold']))

# Data Info

df.info()
```

```python
# 1. Sepal scatter visualization

sb.scatterplot('sepal_length',
               'sepal_width',
               data = df,
               hue = 'species',
               palette = 'Set1',
               edgecolor = 'b',
               s = 150, 
               alpha = 0.7
              )
plt.title('SEPAL LENGTH / SEPAL WIDTH')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend(loc = 'upper left', fontsize = 12)
#plt.savefig('sepal.png')
```

The Setosa species tends to have shorter sepal lenght but a longer sepal width compared to the other 2 species. A majority of the Versicolor and Virginica records overlap in sepal length however there are a good portion of Virginica records that have longer sepal lengths. Takeaways:

    - The Setosa species is generally has a sepal width of at least 3.0 and a sepal length under 5.5
    - The Versicolor species generally has a sepal width under 3.0 and a sepal length between 5.0 and 7.0
    - The Verginica species generally has a sepal width between 3.5 and 2.5 and a sepal length between 5.5 and 8.0

```python
# 2. Petal scatter visualization

sb.scatterplot('petal_length',
               'petal_width',
               data = df,
               hue = 'species',
               palette = 'Set2',
               edgecolor = 'b',
               s = 150, 
               alpha = 0.7
              )
plt.title('PETAL LENGTH / PETAL WIDTH')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.legend(loc = 'upper left', fontsize = 12)
#plt.savefig('petal.png')
```

There appears be a strong distinction for the Setosa species when it comes to petal length and width. Setosa tend to have smaller petal length and widths compared to verginica and versicolor. The Verginca species tends to have the longest length for petals and widths among the three species. Takeaways:

    - Setosa species generally has a petal length between 1 and 2 and a petal width under 0.5
    - Versicolor species generally has a petal length between 3 and 5 and a petal width between 1.0 and 1.5
    - Virginica species generally has a petal length between 5 and 7 and a petal width between 1.5 and 2.5

```python
# 3. Data Heatmap

df_corr = df.corr()

sb.heatmap(df_corr,
           cmap = 'Blues', 
           annot = True, 
           xticklabels = df_corr.columns.values, 
           yticklabels = df_corr.columns.values
          )
plt.title('Iris Data Heatmap', fontsize = 15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)

#plt.savefig('heatmap.png')
```

There appears to be no correlation between:

    - sepal width and sepal length

There is a weak negative correlation between:

    - sepal width and petal width
    - sepal width and petal length

```python
# 4. Scatter Matrix

sb.pairplot(data = df, 
            hue = 'species', 
            palette = ['Red', 'Blue', 'limegreen']
           )
#plt.savefig('pairplot.png')
```

```python
# 5. Distribution plot

plt.subplot(211)
sb.kdeplot(df['sepal_length'], color = 'r', shade = True, label = 'Sepal Length')
sb.kdeplot(df['sepal_width'], color = 'b', shade = True, label = 'Sepal Width')

plt.subplot(212)
sb.kdeplot(df['petal_length'], color = 'coral', shade = True, label = 'Petal Length')
sb.kdeplot(df['petal_width'], color = 'green', shade = True, label = 'Petal Width')

#plt.savefig('dist.png')
```

    - Sepal width appears to be the only metric that is normally distributed
    - Both petal length and petal width have a bimodal distribution

```python
X_var = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].values
y_var = df['species'].values

print(cl('X variable :', attrs = ['bold']), X_var[:5])
print(cl('Y variable :', attrs = ['bold']), y_var[:5])
```

```python
X_var = StandardScaler().fit(X_var).transform(X_var.astype(float))
print(cl(X_var[:5], attrs = ['bold']))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_var, y_var, test_size = 0.20, random_state = 1002)

print(cl('Train set shape :', attrs = ['bold']), X_train.shape, y_train.shape)
print(cl('Test set shape :', attrs = ['bold']), X_test.shape, y_test.shape)
```

```python
k = 3

neigh = KNeighborsClassifier(n_neighbors = k)
neigh.fit(X_train, y_train)

print(cl(neigh, attrs = ['bold']))
```

```python
yhat = neigh.predict(X_test)

print(cl('Prediction Accuracy Score (%) :', attrs = ['bold']), round(accuracy_score(y_test, yhat)*100, 2))
```
